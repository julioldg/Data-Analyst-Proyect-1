## Data Normalization in Python
Uniform the features value with different range. Fair comparisson between different features

Methods of Normalizing data

1) X<sub>new</sub> = X<sub>old</sub>/X<sub>max</sub> -> Simple feature Scaling c [0,1]
2) X<sub>new</sub> = (X<sub>old</sub> - X<sub>min</sub>)/(X<sub>max</sub> - X<sub>min</sub>) -> Min-Max contained [0,1]
3) x<sub>new</sub> = (x<sub>old</sub> - μ) / σ -> Z-score on standard deviation contained [-3, 3]




```python
# Simple feature scaling
df["length"] = df["length"]/df["length"].max()

# Min-max scaling
df["length"] = (df["length"] - df["length"].min()) / (df["length"].max() - df["length"].min())

# Z-Score 
df["length"] = (df["length"] - df["length"].mean()) / (df["length"].std())
```

## Binning in Python

Binning is grouping values into bins

Convert numeric into cathegorical variables

Group a set of numerical values into a set of bins

```python
df["price"] = pd.to_numeric(df["price"], errors="coerce")

bins = np.linspace(min(df["price"]), max(df["price"]), 4)

group_names = ["Low", "Medium", "High"]

df["price-binned"] = pd.cut(df["price"], bins, labels = group_names, include_lowest = True)

df.head(5)
```

## Correlation-Statistics

Pearson Correlation

+ Measure of the strangth of the correlation between two features

    + Correlation coefficient
        
        + Close to +1 Large + relation
        + Close to -1 Large - relation
        + Close to 0, no relationship
    
    + P-value
        
        + P-value < 0.001 Strong
        + P-value < 0.05 Moderate
        + P-value < 0.1 Weak
        + P-value > 0.1 No centrality
    
    + Strong Correlation 
        
        + Correlation coefficient close to 1 or -1
        + P-Value less than 0.001


```python
df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

df = df.dropna(subset=["horsepower", "price"])

from scipy import stats 

pearson_coef, p_value = stats.pearsonr(df["horsepower"], df["price"])

print(pearson_coef)

print(p_value)
```

### Chi-Square: Test for Association

Categorical variables

We use Chi-Square test for association

The test is intended to test how likely it is that an observed distribution is due to a chance

Chi Square test a null hypotesis that the variables are independent

Is there an association between fuel-type and aspiration?

```python
cont_table = pd.crosstab(df["fuel-type"], df["aspiration"])

scipy.stats.chi2_contingency(cont_table, correction = True)
```

El valor p es 5.327871664715056e-08, que es extremadamente pequeño. Esto sugiere que hay una asociación significativa entre las variables categóricas en la tabla de contingencia. Dado que el valor p es mucho menor que 0.05, se puede rechazar la hipótesis nula y concluir que las variables están asociadas.
